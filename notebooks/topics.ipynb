```python
import ast
def print_super_sub_topics(df: pd.DataFrame):
    """Print each SuperTopic with its SubTopics and comments cleanly."""
    grouped = df.groupby("SuperTopicName")

    for super_topic, group in grouped:
        print(f"\nüü¶ SuperTopic: {super_topic} (Total: {group['Frequency'].sum()})")
        
        for _, row in group.iterrows():
            subtopic = row["SubTopicName"] if pd.notna(row["SubTopicName"]) else "Unnamed"
            print(f"   ‚îî‚îÄ‚îÄ üîπ SubTopic: {subtopic}  |  Frequency: {row['Frequency']}")
            
            comments = row["Comments"]
            comments = ast.literal_eval(comments)
            for comment in comments:
                print(f"       üó®Ô∏è  {comment}")
```
